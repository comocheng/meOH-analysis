```python
# Test with df_1 

df = df_1

# List of species: CO, CO2, H2, CH3OH, H20
species_list = df.columns[4:9]
color_list = ['purple','blue','green', 'chocolate','red']

# Assign each species a color
species_dict = dict(zip(species_list, color_list))


fig, ax = plt.subplots(1,5, figsize=(20,4))


# Loop through each temperature and each species and plot mole frac vs. pressure

for n in range(5):
    row_start = n * 6
    for species in species_dict:
        df.iloc[row_start:row_start+6].plot.scatter(x="p (bar)", y=species, c=species_dict[species], ax=ax[n])
        ax[n].autoscale(enable=True, axis='y')
        ax[n].set_title(f"T = {df['T (K)'][row_start]}")
        
    plt.ylabel('Mole fraction')

plt.legend(species_dict.keys())
```

```python
# MOLE FRACTION VS. PRESSURE
# Plot of Mole fraction vs. Pressure for each feed conditions in a 5x5 subplot

fig, ax = plt.subplots(5, 5, sharey=True, figsize=(20,20))


# Loop through each feed, then temperature and each species and plot mole frac vs. pressure

for i in range(len(df_list)):
    for n in range(5):
        row_start = n * 6
        for species in species_dict:
            df_list[i].iloc[row_start:row_start+6].plot.scatter(x="p (bar)", y=species, c=species_dict[species], ax=ax[i,n])
            ax[i,n].autoscale(enable=True, axis='y')
            ax[0,n].set_title(f"T = {df['T (K)'][row_start]}")

#plt.legend(species_dict.keys())
plt.legend(species_dict.keys(), bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
fig.suptitle('Mole fraction vs. pressure', y=0.92)
plt.setp(ax[:,0], ylabel = 'mole fraction')

plt.show()
```
